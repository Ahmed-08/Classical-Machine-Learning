```python
import pandas as pd
import numpy as np

people = pd.read_csv("G:/Dataset/people.csv")
cars = pd.read_csv("G:/Dataset/cars.csv")
appart = pd.read_csv("G:Dataset/appartments.csv")
people.head(3)
```

```python
print("people ", people.shape)
print("cars ", cars.shape)
print("appart ", appart.shape)
```

```python
print("people: ", people.shape[0],"\n", people.isnull().sum())
print()
print("cars: ", cars.shape[0], "\n", cars.isnull().sum())
print()
print("appart: ", appart.shape[0], "\n", appart.isnull().sum())
```

```python
new_cars = cars.fillna(cars.mean())
new_appart = appart.fillna(appart['Rooms'].median())
new_people = people.fillna(people['Height'].median())
```

```python
print(new_cars.isnull().sum())
print()
print(new_appart.isnull().sum())
print()
print(new_people.isnull().sum())
```

```python
transf = np.nan_to_num(cars.loc[:, 'Power'])
availability_of_car = pd.Series(transf.astype(bool))
cars['availability_of_car'] = availability_of_car
cars
```
